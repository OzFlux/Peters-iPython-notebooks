```python
name_ofqc = "../Sites/Whroo/Data/Portal/Whroo_L6.nc"
ds_ofqc = qcio.nc_read_series(name_ofqc)
ldt_ofqc = numpy.array(ds_ofqc.series["DateTime"]["Data"])
fsd_ofqc,f,a = qcutils.GetSeriesasMA(ds_ofqc,"Fsd")
fsdsyn_ofqc,f,a = qcutils.GetSeriesasMA(ds_ofqc,"Fsd_syn")
sa,f,a = qcutils.GetSeriesasMA(ds_ofqc,"solar_altitude")
gpp_ofqc,f,a = qcutils.GetSeriesasMA(ds_ofqc,"GPP_LT")
nee_ofqc,f,a = qcutils.GetSeriesasMA(ds_ofqc,"NEE_LT")
er_ofqc,f,a = qcutils.GetSeriesasMA(ds_ofqc,"ER_LT")
```

```python
name_rep = "../Sites/Whroo/Data/REddyProc/Whroo_L6.nc"
ds_rep = qcio.nc_read_series(name_rep)
ldt_rep = ds_rep.series["DateTime"]["Data"]
fsd_rep,f,a = qcutils.GetSeriesasMA(ds_rep,"Fsd")
gpp_rep,f,a = qcutils.GetSeriesasMA(ds_rep,"GPP")
```
